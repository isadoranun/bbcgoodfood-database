```python
url1 = 'https://www.bbcgoodfood.com/recipes/parkin'
scraper = scrape_me(url1)
```

```python
def scrape_recipe(url):
    try:
        sleep(randint(1,5))
        scraper = scrape_me(url)
        x = {
          "title": scraper.title(),
          "url": url,
        #   "id": url.split('recipes/')[1].split('/')[0],
          "ingredients": scraper.ingredients(),
          "parsed_ingredients": scraper.parsed_ingredients(),
          "nutritional": scraper.nutritional(),
          "keywords": scraper.keywords(),
          "categories": scraper.categories(),
          "description": scraper.description(),
          "instructions": scraper.instructions(),
          "total_time": scraper.total_time()
        }
        
        if len(x['ingredients']) == 0:
            raise Exception("no ingredients parsed for this recipe")
            
        json_data = {
            url: x
        }
        json.dumps(json_data)

        with open('./bbcgoodfood/'+scraper.title()+'.json', 'w') as fp:
            json.dump(json_data, fp, sort_keys=True, indent=4)
        print(scraper.title())
    except Exception as e:
        print("No recipe found for " + url, "|", e)
```

```python
scrape_recipe(url1)
```

```python
from xml.dom import minidom

# parse an xml file by name
mydoc = minidom.parse('sitemap.xml')

items = mydoc.getElementsByTagName('loc')
```

```python
urls = []
for i in items:
    urls.append(i.firstChild.data)
```

```python
import multiprocessing

multiprocessing.cpu_count()
```

```python
from multiprocessing import Pool
import numpy as np

# Run this with a pool of 5 agents having a chunksize of 3 until finished
agents = multiprocessing.cpu_count() - 1
chunksize = np.floor(len(urls) / agents)
with Pool(processes=agents) as pool:
    pool.map(scrape_recipe, urls)

```
